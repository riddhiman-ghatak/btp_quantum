```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from tqdm import tqdm
from neural_spine import SpineNet
```

```python
# Load the CSV file
df = pd.read_csv("/content/btp_data..csv")
X = torch.tensor(df["X"].values, dtype=torch.float32).unsqueeze(1)
```

```python
df.head()
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(1, 32),
            nn.Tanh(),
            nn.Linear(32, 64),
            nn.Tanh(),
            nn.Linear(64, 128),
            nn.Tanh(),
            nn.Linear(128, 64),
            nn.Tanh(),
            nn.Linear(64, 32),
            nn.Tanh(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.model(x)
```

```python
def train_model(y):
    y = torch.tensor(y.values, dtype=torch.float32).unsqueeze(1)
    model = MLP()
    print("Model Architecture:\n", model)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=50, verbose=True)

    for epoch in tqdm(range(1000), desc="Training Progress"):
        optimizer.zero_grad()
        output = model(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()
        scheduler.step(loss)

    return model
```

```python
models = {}
for col in ["Y1", "Y2", "Y3"]:
    print(f"Training model for {col}")
    models[col] = train_model(df[col])
```

```python
# Save weights and biases
def save_model_parameters(model, filename):
    with open(filename, "w") as f:
        for name, param in model.named_parameters():
            f.write(f"{name}: {param.data.tolist()}\n")

for col, model in models.items():
    save_model_parameters(model, f"{col}_params.txt")
```

```python
# Actual vs Predicted Plot
def plot_actual_vs_predicted(col, model):
    with torch.no_grad():
        predictions = model(X).numpy()

    plt.figure(figsize=(8, 5))
    plt.scatter(df["X"], df[col], label=f"{col} Actual", color='blue')
    plt.scatter(df["X"], predictions, label=f"{col} Predicted", color='red')
    plt.xlabel("X")
    plt.ylabel(f"{col} Values")
    plt.title(f"Actual vs Predicted for {col}")
    plt.legend()
    plt.grid()
    plt.show()

for col, model in models.items():
    plot_actual_vs_predicted(col, model)
```
